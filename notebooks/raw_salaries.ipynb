```python
import numpy as np
import pandas as pd
```

```python
df_salaries_og = pd.read_csv("../data/raw/salaries.csv")
```

```python
# Display first few rows of the dataframe
df_salaries_og
```

```python
# Count unique values in each column
df_salaries_og.nunique()
```

```python
# Display datatypes of columns
df_salaries_og.dtypes
```

```python
# Check for missing values in each column
df_salaries_og.isna().any()
```

```python
# Count missing values in each column
df_salaries_og.isna().sum()
```

```python
# Check for duplicated rows
df_salaries_og.duplicated().any()
```

```python
# Drop columns I am not going to need for my analysis
df_salaries_og = df_salaries_og.drop((['nombre', 'idInformacion']), axis = 1)
```

```python
# Drop na's from the columns that had them
df_salaries_og = df_salaries_og.dropna(subset=['denominacion', 'montoneto', 'cargo', 'area', 'montobruto'])
```

```python
# Count missing values in each column
df_salaries_og.isna().sum()
```

```python
# New column for year
df_salaries_og['periodoreportafin'] = pd.to_datetime(df_salaries_og['periodoreportafin'], errors = 'coerce')
```

```python
# New column for year
df_salaries_og['year'] = pd.DatetimeIndex(df_salaries_og['periodoreportafin']).year
```

```python
# Display the dataframe
df_salaries_og
```

```python
# Count unique values in each column
df_salaries_og['year'].unique()
```

```python
# Drop na's from the column year
df_salaries_og = df_salaries_og.dropna(subset=['year'])
```

```python
# Count missing values in each column
df_salaries_og.isna().sum()
```

```python
# Display datatypes of columns
df_salaries_og.dtypes
```

```python
# Transform column year in int
df_salaries_og['year'] = df_salaries_og['year'].astype(int)
```

```python
# Drop columns I am not going to need for my analysis
df_salaries_og = df_salaries_og.drop((['periodoreportainicio', 'periodoreportafin']), axis = 1)
```

```python
# Display the dataframe head
df_salaries_og.head()
```

```python
# Explore my data 
df_salaries_og['new_column'] = df_salaries_og['denominacion'].equals(df_salaries_og['cargo'])
```

```python
# Finding if two columns are the same
df_salaries_og['new_column'].unique()
```

```python
# Drop columns I am not going to need for my analysis
df_salaries_og = df_salaries_og.drop((['denominacion', 'new_column']), axis = 1)
```

```python
#Reset the index
df_salaries_og = df_salaries_og.reset_index(drop=True)
```

```python
# Display the dataframe
df_salaries_og
```
